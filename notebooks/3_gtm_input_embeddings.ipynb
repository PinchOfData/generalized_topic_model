```python
import pandas as pd
df = pd.read_csv('../data/hansard_archive/hansard_speeches_processed.csv', nrows = 1000)

import os
os.environ["TOKENIZERS_PARALLELISM"] = "false" # to avoid some warnings

import sys
sys.path.append('../gtm/')
from corpus import GTMCorpus
from gtm import GTM

# Create a GTMCorpus object
train_dataset = GTMCorpus(
    df, 
    embeddings_type = "SentenceTransformer",
    sbert_model_to_load = "all-MiniLM-L6-v2",
    prevalence = "~ party", 
    content = "~ party",
    max_seq_length = 256
)
```

```python
# Train the model
tm = GTM(
    train_dataset, 
    n_topics=20,
    encoder_input = 'embeddings',
    encoder_hidden_layers = [128,64],
    num_epochs=100
)
```
